```python
df = pd.read_csv("data-till-8th.csv")
df.head()
```

```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import json
plt.style.use('seaborn-pastel')
%matplotlib inline
```

```python
df = df.dropna(axis=1)
```

```python
df.head()
```

```python
import time
import random
```

```python
def random_fmt():
    markers = [',', '+', '.', 'o', '*', 'D', 'p']
    line_styles = ['-', '--', '-.', ':']
    colors = ['b', 'g', 'r', 'c', 'm', 'y']
    
    return random.choice(markers)+random.choice(line_styles)+random.choice(colors)
    
```

```python
def draw_graph1():
    states = df.columns.values[2:]
    plt.figure(figsize=(20,20))
    date_col = [df["date"][i][-4:] for i in range(df.shape[0])]
    for state in states:
        clean_state_data = [json.loads(df[state][i].replace("'", '"'))[0]['confirmed'] for i in range(df.shape[0])]
        plt.plot(date_col, clean_state_data, random_fmt(), label=state)
    
    plt.legend(prop={'size':15})
    plt.yscale("log")
    plt.show()
    time.sleep(.2)
        
```

```python
draw_graph1()
```
