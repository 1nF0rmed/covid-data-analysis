```python
df = pd.read_csv("data-till-8th.csv")
df.head()
```

```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
import json
plt.style.use('seaborn-pastel')
%matplotlib inline
```

```python
df = df.dropna(axis=1)
```

```python
df.head()
```

```python
import time
import random
```

```python
def random_fmt():
    markers = [',', '+', '.', 'o', '*', 'D', 'p']
    line_styles = ['-', '--', '-.', ':']
    colors = ['b', 'g', 'r', 'c', 'm', 'y']
    
    return random.choice(markers)+random.choice(line_styles)+random.choice(colors)
    
```

### Analysis of confirmed cases

```python
def draw_graph1():
    states = df.columns.values[2:]
    plt.figure(figsize=(20,20))
    date_col = [df["date"][i][-4:] for i in range(df.shape[0])]
    for state in states:
        clean_state_data = [json.loads(df[state][i].replace("'", '"'))[0]['confirmed'] for i in range(df.shape[0])]
        plt.plot(date_col, clean_state_data, random_fmt(), label=state)
    
    plt.legend(prop={'size':15})
    plt.yscale("log")
    plt.show()
        
```

```python
draw_graph1()
```

### Analysis of New Cases

```python
new_cases_df = df.copy()
for state in new_cases_df.columns.values[2:]:
    new_cases_df[state] = new_cases_df[state].apply(lambda x:json.loads(x.replace("'", '"'))[0]['confirmed'] )
new_cases_df.head()
```

```python
for state in new_cases_df.columns.values[2:]:
    new_cases_df[state] = new_cases_df[state].diff().fillna(value=0)
new_cases_df.head()
```

```python
def draw_graph2(_df):
    states = _df.columns.values[2:]
    plt.figure(figsize=(20,20))
    date_col = [_df["date"][i][-4:] for i in range(_df.shape[0])]
    for state in states:
        clean_state_data = [_df[state][i] for i in range(_df.shape[0])]
        plt.plot(date_col, clean_state_data, random_fmt(), label=state)
    
    plt.legend(prop={'size':15})
    plt.show()
```

```python
draw_graph2(new_cases_df)
```

```python
growth_df = df.copy()
for state in growth_df.columns.values[2:]:
    growth_df[state] = growth_df[state].apply(lambda x:json.loads(x.replace("'", '"'))[0]['confirmed'] ).pct_change()*100
    growth_df[state] = growth_df[state].fillna(value=0).replace(np.inf, 0.0)
    
growth_df = growth_df.drop('Unnamed: 0',1)
growth_df.head()
```

#### Top 5 States by average growth rate in the last month

```python
growth_df.mean(axis=0).sort_values(ascending=False).nlargest(5)
```

```python
draw_graph2(growth_df)
```

#### Growth Rates in states with metropolitan areas

```python
metro_states = ["Delhi", "Karnataka", "Maharashtra", "West Bengal", "Andhra Pradesh", "Tamil Nadu"]
draw_graph2(growth_df[["date"]+metro_states])
```

# Total Cases Analysis

```python
total_df= df.copy()

for state in df.columns.values[2:]:
    total_df[state] = df[state].apply(lambda x:json.loads(x.replace("'", '"'))[0]['confirmed'] )
    
col_list=list(total_df)
col_list.remove("date")
col_list.remove("Unnamed: 0")
total_df["total_confirmed"]=total_df[col_list].sum(axis=1)

for state in df.columns.values[2:]:
    total_df[state] = df[state].apply(lambda x:json.loads(x.replace("'", '"'))[0]['recovered'] )
total_df["total_recovered"]=total_df[col_list].sum(axis=1)

for state in df.columns.values[2:]:
    total_df[state] = df[state].apply(lambda x:json.loads(x.replace("'", '"'))[0]['deaths'] )
total_df["total_deaths"]=total_df[col_list].sum(axis=1)

total_df.tail()
```

```python
def draw_graph3(_df,col):
    plt.figure(figsize=(15,8))
    date_col = [_df["date"][i][-4:] for i in range(_df.shape[0])]
    clean_col_data = [_df[col][i] for i in range(_df.shape[0])]
    plt.plot(date_col, clean_col_data, random_fmt(), label=col)
    plt.legend(prop={'size':15})
    plt.show()
```

```python
draw_graph3(total_df,"total_confirmed")
```

```python
draw_graph3(total_df,"total_recovered")
```

```python
draw_graph3(total_df,"total_deaths")
```
